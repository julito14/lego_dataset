```python
#df4 --> dataframe wihtout NaN values in set_num and filtered to licensed records only

licensed_sorted = df4[['year', 'parent_theme']].sort_values('year')
licensed_sorted['count_col'] = 1
licensed_sorted.head()

```

```python
licensed_sorted_gb = licensed_sorted.groupby(['year', 'parent_theme'])
```

```python
max_df = licensed_sorted_gb.count().sort_values('count_col', ascending=False)
```

```python
max_df = max_df.reset_index()
```

```python
max_df_top = max_df.drop_duplicates(['year'])
#in order to get the maximum number of set releases per year, we can apply the droo_duplicates method (based on the year).
# what will happen is that only the first record that appears will be preserved. The next ones, of same year, will be dropped. So, if the dataset is ordered by the number of releases from highest to lowest, then we will only keep the highest number of releases per year !


```

```python
max_df_top=max_df_top.sort_values('year')
```

```python
max_df_top
```

```python
new_era = int(max_df_top[max_df_top['year']==2017]['year'])
```
