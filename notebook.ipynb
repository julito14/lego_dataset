```python
# Use this cell to begin your analyses, and add as many cells as you would like!
import numpy as np
import pandas as pd
```

```python
df_sets = pd.read_csv('datasets/lego_sets.csv')
```

```python
#cleaning the set_num column.
df_sets = df_sets.dropna(subset=['set_num'])
```

```python
df_sets.set_num.isna().unique()
```

```python
#Load the parent theme dataset
```

```python
df_parent = pd.read_csv('datasets/parent_themes.csv')
df_parent.head()
```

```python
s = df_parent.name.unique()
np.sort(s)
```

```python
df3 = pd.merge(df_sets, df_parent, left_on='parent_theme', right_on='name')
```

```python
df3.head()
```

```python
#verifying that set_num values are unique...
df3.set_num.nunique()
df3.shape


df3.is_licensed.value_counts()
#we see there are at least 1k sets that are licensed. 
```

```python
#Let's filter the df to get only set that are licensed...
df4 = df3[df3['is_licensed']==True]
```

```python
#df4 contains only sets that have been licensed...
df4
```

```python
#determining the total number of sets that have been licensed...

df4.shape
```

```python
n_licensed_sets = df4.set_num.count()
```

```python
n_licensed_sets
```

```python
df4.parent_theme.value_counts()
#we see that there is only one 'Star Wars' unique value! So we don't have to clean this data. 
```

```python
df5 = df4[df4['parent_theme']=='Star Wars']
```

```python
n_licensed_sets_SW = df5.set_num.count()
```

```python
the_force = (n_licensed_sets_SW/n_licensed_sets)*100
```
